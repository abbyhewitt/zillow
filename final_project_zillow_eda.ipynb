```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## got errors of mixed data types for columns,so designated them as strings
data16 = pd.read_csv('zillow_data/properties_2016.csv', dtype = {
    'hashottuborspa': str,
    'propertycountylandusecode':str,
    'propertyzoningdesc':str,
    'fireplaceflag': str,
    'taxdelinquencyflag': str,
})
```

```python
data16.describe()
```

```python
data16.shape
```

```python
train_data16 = pd.read_csv('zillow_data/train_2016_v2.csv')
```

```python
train_data16.shape
```

```python
train_data16.head()
```

```python
#merge the 2016 sales data and property data on parcelid
sales_16 = pd.merge(data16, train_data16, how = 'inner', on = 'parcelid')
```

```python
sales_16.shape
```

```python
#drop columns that have fewer than 10,000 values
sales_16 = sales_16.drop(sales_16.columns[sales_16.apply(lambda col: col.count() < 10000)], axis=1)
```

```python
sales_16.shape
```

```python
sales_16.describe()
```

```python
sales_16.count()
```

```python
#drop redundant columns and columns that are not relevant to price
sales_16 = sales_16.drop(columns = ['parcelid','airconditioningtypeid','calculatedbathnbr','finishedsquarefeet12','fips',
'fullbathcnt','garagecarcnt','latitude','longitude', 'pooltypeid7', 'propertycountylandusecode',
'rawcensustractandblock','threequarterbathnbr','assessmentyear','censustractandblock','roomcnt'], axis=1)
```

```python
sales_16.shape
```

```python
sales_16.describe()
```

```python
from scipy.stats.mstats import winsorize
```

```python
#remove outliers
sales_16['bathroomcnt'] = winsorize(sales_16['bathroomcnt'], (None, 0.01))
sales_16['bedroomcnt'] = winsorize(sales_16['bedroomcnt'], (None, 0.01))
sales_16['calculatedfinishedsquarefeet'] = winsorize(sales_16['calculatedfinishedsquarefeet'], (None, 0.01))
sales_16['garagetotalsqft'] = winsorize(sales_16['garagetotalsqft'], (None, 0.01))
sales_16['lotsizesquarefeet'] = winsorize(sales_16['lotsizesquarefeet'], (None, 0.01))
sales_16['unitcnt'] = winsorize(sales_16['unitcnt'], (None, 0.01))
```

```python
sales_16.describe()
```

```python
year_built = sales_16.groupby('yearbuilt')
```

```python
year_built['structuretaxvaluedollarcnt'].mean().plot();

plt.title("Average Assessed Value Over Year Built")
plt.xlabel("Year Built")
plt.ylabel("Assessed Value")
plt.show()

```

```python
bedroom_count = sales_16.groupby('bedroomcnt')
bedroom_count['structuretaxvaluedollarcnt'].mean().plot(kind="bar");

plt.title("Average Assessed Value per Bedroom Count")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Assessed Value")
plt.show()
```

```python
#change transaction date to datetime format
from datetime import date, datetime
sales_16['transaction_datetime'] = pd.to_datetime(sales_16['transactiondate'])
sales_16['transaction_datetime']
```

```python
## add column to group transactions by week
sales_16['transaction_week'] = sales_16['transaction_datetime'].map(lambda x: x.isocalendar()[1])
## plot weekly sales volume 
transaction_date = sales_16.groupby('transaction_week')
transaction_date['transaction_week'].count().plot(kind = "bar", rot = 90, figsize = (10,8))
plt.xlabel("Week of Year")
plt.ylabel("Transactions")
plt.title("2016 Transactions by Week")
plt.show()
```

```python
#Significant drop off after 41st week of the year.
```

```python
## plot log error by week
transaction_date['logerror'].mean().plot(kind = "bar", rot = 90, figsize = (10,8))
plt.xlabel("Week of Year")
plt.ylabel("Avg. Log Error")
plt.title("2016 Avg Log Error by Week")
plt.show()
```
